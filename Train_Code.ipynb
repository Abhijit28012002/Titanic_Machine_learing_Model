```python
def  set_age(row):
    
    Pclass = row[0]
    age = row[1]
    
    if not age:
        if Pclass == 1:
            return 40
        elif Pclass == 2:
            return 28
        else:
            return 24
        
    else:
        return age
```

```python
dataset[['Pclass','Age']].apply(set_age,axis=1)
```
